#### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

#### Load Dataset

```python
current_dir = os.getcwd()
titanic = pd.read_csv('../Data/titanic.csv')
titanic.head()
```

#### Inspect Data

```python
#display data types
titanic.info()
```

```python
#check statistics summary
titanic.describe()
```

#### Check/handle missing values (for variables in hypotheses only)

```python
titanic.isna().sum()
```

```python
#compute median age
median_age = titanic['Age'].median()
#impute missing data for age
titanic['Age']= titanic['Age'].fillna(median_age)
#check for missing values in age column
titanic['Age'].isna().sum()
```

#### Hypotheses to Check:

* Determine if the survival rate is associated to the class of passenger
* Determine if the survival rate is associated to the gender
* Determine the survival rate is associated to the age

#### Univariate Analysis

```python
#set up dimensions of histogram
plt.figure(figsize=(5, 4))
#create countplot by passenger class
sns.countplot(x ='Pclass', data =titanic, palette='deep', hue ='Pclass')
plt.title('Count of Passengers by Class')
plt.ylabel('Number of Passengers')
plt.xlabel('Passenger Class')
plt.xticks([0, 1, 2], ['1st Class', '2nd class', '3rd class'])
plt.savefig('../Results/classCount.png')
plt.show()
```

```python
#create simple dictionary for palette
palette = {'male': 'blue', 'female': 'pink'}
#set up dimensions of histogram
plt.figure(figsize=(5, 4))
#create countplot by gender
sns.countplot(x = 'Sex', data = titanic, palette=palette, hue='Sex')
plt.title('Count of Passengers by Gender')
plt.ylabel('Number of Passengers')
plt.xlabel('Gender')
plt.xticks([0, 1], ['Male', 'Female'])
plt.savefig('../Results/genderCount.png')
plt.show()
```

```python
#set up dimensions of histogram
plt.figure(figsize=(5, 4))
sns.histplot(titanic['Age'], bins = 10, color='lightgreen')
# Add labels and title from the matplotlib library
plt.ylabel("Frequency of Age")
plt.title("Histogram of Passenger Ages")
plt.savefig('../Results/ageHistogram.png')
```

#### Bivariate Analysis

```python
#calculate survival rates by class
survival_rates_class = titanic.groupby('Pclass')['Survived'].mean()
print((survival_rates_class*100).round(2).astype(str)+'%')
#adjust size of graph
plt.figure(figsize=(5, 4))
#create the bar plot for survival rates by class
survival_rates_class.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Survival Rates by Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.xticks(ticks=[0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'], rotation=0)
plt.ylim(0, 1)  
plt.grid(axis='y')
plt.savefig('../Results/survivalByClass.png')
plt.show()
```

```python
#calculate survival rates by gender
survival_rates_gender = titanic.groupby('Sex')['Survived'].mean() 
print((survival_rates_gender*100).round(2).astype(str)+'%')
#adjust size of graph
plt.figure(figsize=(5, 4))
#create the bar plot for survival rates by gender
survival_rates_gender.plot(kind='bar', color=['pink', 'blue'])
plt.title('Survival Rates by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.ylim(0, 1)  
plt.grid(axis='y')
plt.savefig('../Results/survivalByGender.png')
plt.show()
```

```python
#divide ages into age-groups by creating age bins
age_bins = [0, 18, 41, 61, 81]  
age_labels = ['0-17', '18-40', '41-60', '61-80'] 

#create new'AgeGroup' column to categorize ages
titanic['AgeGroup'] = pd.cut(titanic['Age'], bins=age_bins, labels=age_labels, right=False)

#calculate survival rates for each age group
survival_rate_by_age = titanic.groupby('AgeGroup')['Survived'].mean() * 100
print(survival_rate_by_age.round(2))

#create bar chart rate by age group
plt.figure(figsize=(5, 4))
survival_rate_by_age.plot(kind='bar', color='skyblue')
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate (%)')
plt.xticks(rotation=0)
plt.savefig('../Results/survivalByAgeGroup.png')
plt.show()


```

```python
#set up dimensions of histogram
plt.figure(figsize=(5, 4))

#create a histogram for age of both survived and not survived
sns.histplot(data=titanic, x='Age', bins=20, hue='Survived', multiple='stack', palette='pastel')

#edit histogram labels
plt.title('Histogram of Age by Survival Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xlim(0, 80)  # Limit age range for better visualization
plt.legend(title='Survival Status', labels=['Not Survived', 'Survived'])
plt.savefig('../Results/survivalByAge.png')
plt.show()
```
